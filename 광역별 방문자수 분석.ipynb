```python
kt_2021 = pd.read_csv('방문자수_KT/2021_kt광역별 방문자 수.csv', encoding='cp949')
kt_2022 = pd.read_csv('방문자수_KT/2022_kt광역별 방문자 수.csv', encoding='cp949')
kt_2023 = pd.read_csv('방문자수_KT/2023_kt광역별 방문자 수.csv', encoding='cp949')
```

```python
skt_2021 = pd.read_csv('방문자수_SKT/2021_skt광역별 방문자 수.csv', encoding='cp949')
skt_2022 = pd.read_csv('방문자수_SKT/2022_skt광역별 방문자 수.csv', encoding='cp949')
skt_2023 = pd.read_csv('방문자수_SKT/2023_skt광역별 방문자 수.csv', encoding='cp949')
```

```python
kt_2021['연도'] = 2021
kt_2022['연도'] = 2022
kt_2023['연도'] = 2023
skt_2021['연도'] = 2021
skt_2022['연도'] = 2022
skt_2023['연도'] = 2023
```

```python
df = pd.concat([kt_2021, kt_2022, kt_2023, skt_2021, skt_2022, skt_2023], ignore_index=True)
```

```python
df
```

```python
df.loc[:,'방문자수'] = df['광역지자체 방문자 수'] + df['기초지자체 방문자 수']
df
```

```python
df = df[['광역지자체명','기초지자체명','연도','방문자수']]
df['기초지자체명'] = df['기초지자체명'].str.slice(start=0,stop=4)
df['광역지자체명'] = df['광역지자체명'].str.slice(start=0,stop=2)
df
```

```python
df.loc[df['광역지자체명'] == '전북','광역지자체명'] = '전라'
df.loc[df['기초지자체명'] == '세종특별','기초지자체명'] = '특별자치시'
```

```python
df = df.groupby(['광역지자체명','기초지자체명','연도'])['방문자수'].sum().reset_index()
```

```python
df['위치'] = df['광역지자체명'] + df['기초지자체명']
df['위치'] = df['위치'].str.strip()
```

```python
df = df[['연도','위치','방문자수']]
df
```
