```python
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# 한글설정 & 마이너스표시
plt.rc('font',family='Malgun Gothic') 
plt.rc('axes',unicode_minus=False)
```

```python
df = pd.read_excel('korea_administrative_division_latitude_longitude.xlsx')
df
```

```python
df['do'] = df['do'].str.slice(start=0,stop=2)
df
```

```python
df['do_city'] = df['do'] + df['city']
df['docity'] = df['do_city']
df
```

```python
df= df.drop(['do','city','do_city'],axis=1)
df
```

```python
df.loc[df['docity'] == '경상군위군','docity'] = '대구군위군'
df.loc[df['docity'] == '경기옹진군','docity'] = '인천옹진군'
df.loc[df['docity'] == '충청남세종로','docity'] = '세종특별시'
```

```python
df.to_excel('경도위도정리.xlsx',index=False)
```
